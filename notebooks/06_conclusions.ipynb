# Conclusions and Future Work

This notebook summarizes the findings of the machine learning pipeline
developed for thermography-based fault detection.

## Study Context

The objective of this project was to evaluate the effectiveness of
machine learning models in diagnosing electrical faults from
thermographic data, complemented with electrical measurements.

After data cleaning and preprocessing, the final dataset was reduced to
two operating conditions: *Normal* and *Contactos flojos*.

## Main Findings

- Exploratory Data Analysis revealed high dimensionality and the presence
  of noise, justifying dimensionality reduction techniques.
- PCA effectively reduced the feature space while preserving most of the
  variance.
- Multiple machine learning models were evaluated, including SVM,
  Random Forest, XGBoost, and CNN.
- Under the reduced dataset conditions, Support Vector Machine achieved
  the lowest Log Loss (0.11), indicating superior probabilistic
  performance.

## Model Selection

Although more complex models such as CNN and XGBoost demonstrated strong
performance in some scenarios, their effectiveness was limited by the
small dataset size.

Support Vector Machine was selected as the final model due to its
robustness and generalization capability in high-dimensional, small-sample
settings.

## Limitations

- The dataset size was limited after removing incomplete samples.
- Class imbalance and reduced operating conditions may affect
  generalization.
- CNN models are likely to suffer from overfitting under such data
  constraints.

## Future Work

Future research could focus on:

- Expanding the dataset with additional thermographic samples.
- Incorporating more electrical parameters to enrich feature space.
- Evaluating deep learning architectures with larger datasets.
- Deploying the trained model in a real-time monitoring system.

## Final Conclusion

This study demonstrates that machine learning techniques can effectively
support technical diagnosis of thermographic data, even under limited
data conditions.

The results highlight the importance of model selection based on dataset
characteristics, with simpler models outperforming complex architectures
when data availability is constrained.
